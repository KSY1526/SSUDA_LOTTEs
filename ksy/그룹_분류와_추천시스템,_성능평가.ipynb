```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(rct_num):
    receipt = new_df[new_df['rct_no'] == rct_num] # receipt : 영수증
    tem = new_df.copy()
    tem['cosine'] = cosine_similarity(receipt.iloc[:,:20].to_numpy().reshape(1,-1), new_df.iloc[:,:20].to_numpy()).T
    
    tem = tem[['rct_no', 'cosine']]
    tem = tem.sort_values(by = ['cosine'], ascending = False)
    #tem = tem.reset_index()
    people = []

    # 코사인 유사도가 높은 상위 5명(tem[0]은 본인이 나옴)
    for i in list(tem['rct_no'][1:6]):
        # 영수증에 맞는 사람 찾기
        people.append(record[record['rct_no'] == i]['cust'].values[0])
    
    product = []
    for i in people:
        product.append(record[record['cust'] == i].sample(n = 1)['pd_c'].values[0])

    # 본인 선택 상품과 일치 여부(성능평가)
    myname = record[record['rct_no'] == rct_num]['cust'][:1].values[0]
    myproduct = set(record[record['cust'] == myname]['pd_c'])
    mycnt = len(myproduct)

    # 상품 이름 출력 위해
    names = []
    overlap = 0
    for i in product:
        if i in myproduct: # 이미 산 제품이라면
            overlap += 1
        else:
            names.append(products[products['pd_c'] == i]['pd_nm'].values[0])
    #print(*names) 
    #print(mycnt, '개 제품 구매했는데,', overlap, '개 겹치게 추천')

    return overlap / mycnt

recommend('A03089375637')
```

```python
def randomrecommend(rct_num):
    myname = record[record['rct_no'] == rct_num]['cust'][:1].values[0]
    myproduct = set(record[record['cust'] == myname]['pd_c'])
    mycnt = len(myproduct)

    overlap = 0
    for i in products['pd_c'].sample(n = 5).values:
        if i in myproduct: # 이미 산 제품이라면
            overlap += 1

    return overlap / mycnt

randomrecommend('A03089375637')
```

```python
score1 = 0
score2 = 0
for i in tqdm(new_df['rct_no'].sample(n = 100).values):
    score1 += recommend(i)
    score2 += randomrecommend(i)

print(score1,score2)
```
