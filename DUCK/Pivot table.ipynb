```python
tem = record.groupby(['pd_c'])['buy_am'].median()
tem
```

```python
def fun(x1, x2):
    if x2 >= tem[x1]:
        return x1 + '1'
    else:
        return x1 + '2'


record['buy'] = record.apply(lambda x : fun(x['pd_c'], x['buy_am']), axis = 1)

record.head()
```

```python
tem = pd.pivot_table(record, index = 'cust', columns = 'buy', values = 'buy_ct', aggfunc = 'sum')
tem.fillna(0, inplace = True)
tem.head()
```
